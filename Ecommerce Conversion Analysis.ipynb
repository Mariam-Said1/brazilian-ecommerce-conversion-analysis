```python
import pandas as pd

orders = pd.read_csv("olist_orders_dataset.csv")
orders.head()
```

```python
order_items = pd.read_csv("olist_order_items_dataset.csv")
order_items.head()
```

```python
orders["order_status"].value_counts()
```

```python
total_orders = len(orders)

shipped_orders = orders[
    orders["order_status"].isin(["shipped", "delivered"])
].shape[0]

delivered_orders = orders[
    orders["order_status"] == "delivered"
].shape[0]

total_orders, shipped_orders, delivered_orders
```

```python
ship_rate = shipped_orders / total_orders
delivery_rate_from_shipped = delivered_orders / shipped_orders
overall_delivery_rate = delivered_orders / total_orders

ship_rate, delivery_rate_from_shipped, overall_delivery_rate
```

```python
products = pd.read_csv("olist_products_dataset.csv")
products.head()
```

```python
orders_items = orders.merge(
    order_items,
    on="order_id",
    how="left"
)

orders_items_products = orders_items.merge(
    products[["product_id", "product_category_name"]],
    on="product_id",
    how="left"
)

orders_items_products.head()
```

```python
category_delivery = (
    orders_items_products
    .groupby("product_category_name")
    .agg(
        total_orders=("order_id", "nunique"),
        delivered_orders=("order_status", lambda x: (x == "delivered").sum())
    )
    .assign(delivery_rate=lambda df: df["delivered_orders"] / df["total_orders"])
    .sort_values("delivery_rate", ascending=False)
)

category_delivery.head(10)
```

```python
category_delivery_fixed = (
    orders_items_products
    .groupby("product_category_name")
    .agg(
        total_orders=("order_id", "nunique"),
        delivered_orders=("order_id", lambda x: x[orders_items_products.loc[x.index, "order_status"] == "delivered"].nunique())
    )
    .assign(delivery_rate=lambda df: df["delivered_orders"] / df["total_orders"])
    .sort_values("delivery_rate", ascending=False)
)

category_delivery_fixed.head(10)
```

```python
tmp = orders_items_products[["order_id", "order_status", "product_category_name"]].dropna()

category_delivery_fixed = (
    tmp.groupby("product_category_name")
    .agg(
        total_orders=("order_id", "nunique"),
        delivered_orders=("order_id", lambda s: s[tmp.loc[s.index, "order_status"].eq("delivered")].nunique())
    )
)

category_delivery_fixed["delivery_rate"] = (
    category_delivery_fixed["delivered_orders"] / category_delivery_fixed["total_orders"]
)

category_delivery_fixed.sort_values("delivery_rate", ascending=False).head(10)
```

```python
import pandas as pd

orders = pd.read_csv("olist_orders_dataset.csv")
order_items = pd.read_csv("olist_order_items_dataset.csv")
products = pd.read_csv("olist_products_dataset.csv")

orders_items_products = (
    order_items
    .merge(orders[["order_id", "order_status"]], on="order_id", how="left")
    .merge(products[["product_id", "product_category_name"]], on="product_id", how="left")
)

orders_items_products.head()
```

```python
tmp = orders_items_products[["product_category_name", "order_id", "order_status"]].dropna()

category_delivery_fixed = (
    tmp.groupby("product_category_name")
    .agg(
        total_orders=("order_id", "nunique"),
        delivered_orders=("order_id", lambda s: s[tmp.loc[s.index, "order_status"].eq("delivered")].nunique())
    )
)

category_delivery_fixed["delivery_rate"] = (
    category_delivery_fixed["delivered_orders"] / category_delivery_fixed["total_orders"]
)

category_delivery_fixed.sort_values("delivery_rate", ascending=False).head(10)
```

```python
failed_orders = orders[
    orders["order_status"].isin(["canceled", "unavailable"])
][["order_id", "order_status"]]

failed_orders.head(), failed_orders.shape
```

```python
failed_revenue = (
    order_items
    .merge(failed_orders, on="order_id", how="inner")
    .groupby("order_status")
    .agg(
        failed_orders=("order_id", "nunique"),
        lost_item_revenue=("price", "sum")
    )
)

failed_revenue["avg_lost_revenue_per_order"] = (
    failed_revenue["lost_item_revenue"] / failed_revenue["failed_orders"]
)

failed_revenue
```
